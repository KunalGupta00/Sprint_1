Objective :

The objective of this dataset is to provide a detailed breakdown of the 2024 General Election results in India by state, constituency, candidate, and party. It can be used to analyze voting patterns, the performance of political parties, and election outcomes across different regions.

GitHub Link

Features :

1. State: Contains the names of the states and Union Territores of India. Total number of unique values are 36 (28 states and 8 union territories).

2. Constituency: Contains the names of the constituency in which the votes have been accounted for. There are 541 unique values in this table, however according to ECI, there are 543 constituencies in India.

3. Candidate: There stood 8101 unique candidates in this year's Lok Sabha Elections. Each candidate was either from a party, or an independent candidate.

4. Party: A total of 746 different parties had their registered candidates in the elections, including Independent candidates and 'None of the above (NOTA)". Most candidates were placed by the Bahujan Samaj Party (488), followed by the Bharatiya Janata Party (441) and Indian National Congress (328).

5. EVM Votes: Number of EVM Votes secured by respective candidates. It contained missing values represented by '-', which were replaced by '0' by me.

6. Postal Votes: Number of postal votes secured by respective candidates.

7. Total Votes: Total number of votes, by adding the postal votes and evm votes.

8. % of total votes: Percentage of votes secured by that candidate, in that constituency.

9. Result: Says whether that candidate won or lost.

Data Visualization

Chart 1

```python
candidates_per_state = data[data['Candidate'] != 'NOTA'].value_counts('State').head(10)
plt.title('Number of Candidates per Each State')
sns.barplot(x=candidates_per_state, y=candidates_per_state.index)
plt.show()
```

```python
parties_per_state = data[~data['Party'].isin(['None of the Above', 'Independent'])].value_counts('State').head(10)

plt.title('Number of Parties per Each State')
sns.barplot(x=parties_per_state, y=parties_per_state.index)
plt.show()
```

```python
Insight:Uttar Pradesh had the highest number of constituencies in the country, followed by Maharashtra and West Bengal.
```

```python
data.groupby(['State']).Constituency.nunique().sort_values().plot.barh(figsize=(14,11))
plt.title("Number of Constituencies per state")
plt.xlabel("Number of constituencies");
```

```python
party_wins = data[data['Result'] == 'Won'].value_counts('Party').head(10)

plt.title('Number of Party Members Won')
sns.barplot(x=party_wins, y=party_wins.index)
plt.show()
```

```python
# Get the top 10 states by total votes
total_votes_per_state = data.groupby('State')['Total Votes'].sum()
total_votes_per_state = total_votes_per_state.sort_values(ascending=False).head(10)

# Convert 'Total Votes' to numeric if necessary
total_votes_per_state = pd.to_numeric(total_votes_per_state)

# Create a pie chart for the total votes per state
plt.figure(figsize=(8, 8))
plt.pie(total_votes_per_state, labels=total_votes_per_state.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 States by Total Votes')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

```python
# Get the top 10 states by total votes
total_votes_per_state = data.groupby('State')['Total Votes'].sum()
total_votes_per_state = total_votes_per_state.sort_values(ascending=True).head(6)

# Convert 'Total Votes' to numeric if necessary
total_votes_per_state = pd.to_numeric(total_votes_per_state)

# Create a pie chart for the total votes per state
plt.figure(figsize=(8, 8))
plt.pie(total_votes_per_state, labels=total_votes_per_state.index, autopct='%1.1f%%', startangle=90)
plt.title('Bottom 5 States by Total Votes')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```
